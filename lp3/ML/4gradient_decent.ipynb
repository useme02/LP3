```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x+3)**2

def g(x):
    return 2*(x+3)
```

```python
x_start=2
lr=0.01
precision=0.000001
max_iter=1000
```

```python
x=x_start
gd=[x]

for i in range (max_iter):
    x_new=x-lr*g(x)
    gd.append(x_new)
    if abs(x-x_new)<precision:
        break
    x=x_new
x_min=x
```

```python
print(f"Local minima of the function y=(x+3)**2 is {f(x_min):.10f} at x= {x_min:.10f}")
```

```python
x_val=np.linspace(-6,1,100)
y_val=f(x_val)

plt.plot(x_val,y_val,label="y=(x+3)**2",color='blue')

plt.scatter(gd,[f(x) for x in gd],color="red",label="Gradient Decent")


plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```
